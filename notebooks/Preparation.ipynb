```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../data/primary_studies.xlsx")
```

```python
df.groupby("Source")["Title"].count()
```

```python
sns.set(font_scale=1.2)
sns.set_theme(style="white", palette=None)
fig, ax = plt.subplots(figsize=(5, 3), dpi=300)
plot = sns.histplot(data=df, x="Publication year", hue="Publication type", multiple="stack", stat="count", discrete=True)
plot.set_ylabel("Publication count")
plot.tick_params(axis='x', rotation=90)
plot.set_xticks(list(range(2000, 2023)))
plot.set_yticks(list(range(0, 20, 2)))
plt.tight_layout()
fig.savefig("../data/publication_count.pdf")
```

```python
df['Main category'] = pd.Categorical(df['Main category'], ['application design', 'self-adaptation', 'sensor usage', 'computation offloading', 'user interface', 'code-level optimization', 'network usage'])

sns.set(font_scale=1.2)
sns.set_theme(style="white", palette=None)
fig, ax = plt.subplots(figsize=(5, 3), dpi=300)
plot = sns.histplot(data=df, y="Main category", stat="count", discrete=True)
plot.set_xlabel("Category count")
plt.tight_layout()
fig.savefig("../data/category_count.pdf")
```
